# Preprocessing

The following code combines dataframes `books_df` and `ratings_df` using the ISBN number. 

```python
books_with_reviews = books_df.merge(ratings_df, on = "ISBN")
books_with_reviews.head()
```

If you look at the output above, you will notice that the dataframe now contains columns `User-ID` and `Book-Rating`. This combined dataframe stores information on each book along with all of its ratings.

```python
# remove all rows that contain missing values
books_with_reviews.dropna(axis = 0, inplace = True)

# drop these columns as well
books_with_reviews.drop(['Year-Of-Publication','Image-URL-S','Image-URL-M'], axis = 1, inplace = True)

# remove all ratings of 0 because we just want to use explicit ratings. See the dataset description on Kaggle for more information.
books_with_reviews.drop(index = books_with_reviews[books_with_reviews['Book-Rating'] == 0].index, inplace = True)

# remove the index column since it is not necessary
books_with_reviews.reset_index(drop = True, inplace = True)

# see the updated data
books_with_reviews.head()
```

```python
matrix = books_with_reviews.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating').fillna(0)
matrix
```

```python
from sklearn.neighbors import NearestNeighbors

knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
knn.fit(matrix)
```

```python
matrix.index
```

```python
def recommend(book):    
    # get 5 other books
    book = book.lower()
    distances, indices = knn.kneighbors(matrix.loc[book].values.reshape(1, -1), n_neighbors=6)
    
    for i in range(1, len(distances.flatten())):
        print(matrix.index[indices.flatten()[i]])
```

```python
book = "Oliver Twist"

recommend(book)
```
